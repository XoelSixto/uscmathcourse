## Rectas y planos

```python
def plot_axis(ax):
    ax.spines['bottom'].set_color('black')
    ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0
    ax.spines['left'].set_color('black')
    ax.spines['left'].set_position(('data',0))
    return 
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

xs = np.arange(-2., 2., 0.1)
a, b = 1., 0.
ys = a*xs+b
ax.plot(xs, ys, color='blue', lw=2)
plot_axis(ax)
ax.set_title('recta')
fig.tight_layout();
```

**Cuestión:** La figura es la gráfica de una recta. ¿Cuál es la ecuación que cumplen los puntos de la recta? ¿Cuál es su pendiente? ¿Cuál es la derivada de la función? ¿Es igual para todos los $x$s?

**Cuestión:** ¿Cual es la ecuación general de la recta?

----

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')

xs = np.linspace(0., 1., 20)
ys = np.linspace(0., 1., 20)

xms, yms = np.meshgrid(xs, ys)
a, b, c = 0., 1., 1. 
zms = a*xms+b*yms+c

cmap = plt.cm.jet
# ax.view_init(0., 45.)
sf = ax.plot_surface(xms, yms, zms, rstride=1, cstride=1, alpha=1., cmap=cmap)
ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$y$', fontsize=16)
ax.set_zlabel('$z$', fontsize=16)
ax.set_zlim([0., 2.])
fig.colorbar(sf, ax=ax);

fig.tight_layout();
```

La figura es un plano. ¿Dónde está el origen? ¿Cual es el valor de $z$ en el origen? Fíjate en $y=0$, ¿es una recta? ¿Qué pendiente tiene? Fíjate en $y=1$, ¿es una recta? ¿qué pendiente tiene? ¿Cuál es su pendiente en la dirección $x$? ¿y en $y$? ¿Cuál es la ecuación de este plano?

## Límites, continuidad

```python
fig, ax = plt.subplots(figsize=(8, 6))

xs = np.arange(-3.*np.pi, 3.*np.pi, 0.2)
ys = np.sin(xs)/xs

ax.plot(xs, ys, color='blue', lw=2)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$f(x)$', fontsize=14)
ax.set_ylim([-0.3, 1.1])
ax.set_title('$\sin(x)/x$');
```

**Cuestión:** La imagen es la gráfica de la función $\sin(x)/x$. ¿La función tienen límite en el origen? ¿Es continua? ¿Cuales son sus derivadas? ¿En qué puntos sus derivadas son nulas?

**Cuestión:** ¿Cuándo la función es nula? ¿Cuáles son los puntos máximos y mínimos?

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

# ax is a Axes3D object
xs = np.linspace(-3.*np.pi, 3.*np.pi, 60)
ys = np.linspace(-3.*np.pi, 3.*np.pi, 60)
xms, yms = np.meshgrid(xs, ys)
rms = np.sqrt(xms ** 2 + yms ** 2)
zms = np.sin(rms)/rms

# cmap = plt.cm.hot
# other colormaps: plt.cm.jet, plt.cm.warmhot, plt.cm.Blues, plt.cm.plasma
cmap = plt.cm.jet
ct = ax.contour(xms, yms, zms, zdir='z', offset=1.2, alpha=0.4, cmap=cmap)
# ct = ax.contour(mxs, mys, mzs, zdir='x', offset=-5, alpha=0.4, cmap=cmap)
# ct = ax.contour(mxs, mys, mzs, zdir='y', offset=4, alpha=0.4, cmap=cmap)
# rstride are the width or the row, cstride the width of the columns (in units!)
sf = ax.plot_surface(xms, yms, zms, rstride=1, cstride=1, alpha=1., cmap=cmap)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
# ax.set_zlim(-1.1, 1.2);
ax.set_title('$\sin(\sqrt{x^2 + y^2})/\sqrt{x^2+y^2}$')
fig.colorbar(sf, ax=ax);
```

**Cuestión:** La imagen es la gráfica en color de la función $\sin(\sqrt{x^2+y^2})/\sqrt{x^2+y^2}$. ¿Tiene límite en el origen? ¿Es continua en el origen? ¿Qué son los discos que aparecen en la parte superior?  

**Cuestión:** ¿Cuáles son los puntos máximos y mínimos en la zona dibujada? ¿Cuántos mínimos hay?

```python
fig, ax = plt.subplots(figsize=(8, 6))

# color fill contour
c0 = ax.contourf(xms, yms, zms, 10, alpha=0.8, cmp=plt.cm.jet)
# lines contour
c1 = ax.contour(xms, yms, zms, 10, colors='black', alpha=0.5, linewidth=.25);
c1.clabel(fontsize=10, inline=1)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
fig.colorbar(c0, ax=ax);
```

**Cuestión:** Esta es la misma gráfica, de la función $\cos(x^2+y^2)$, pero vista desde arriba. ¿A qué puntos corresponde el valor de $z$=0? ¿Qué son las líneas? ¿A qué corresponden?

## ¡Taylor!

```python
fig, ax = plt.subplots(figsize=(8, 6))

xs = np.linspace(0, 3.*np.pi, 200)
ys = np.sin(xs)/xs

ax.plot(xs, ys, color='blue', lw=3)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$f(x)$', fontsize=14)

xs_taylor = np.linspace(0., 3.*np.pi, 6)
ys_taylor = np.sin(xs_taylor)/xs_taylor
ax.plot(xs_taylor, ys_taylor, color='red', lw=3, 
        marker='o', markersize=10)
ax.set_ylim([-0.3, 1.1])
ax.set_title('$\sin(x)/x$');
```

**Cuestión:** La línea azul corresponde a la gráfica de la función $\cos(x)$, mientras que los puntos rojos son valores de dicha función, las rectas rojas une dichos puntos. ¿Qué pasa si en vez de considerar 6 puntos dibujásemos 12? ¿Y 100? ¿Crees que la curva se podría dar como una sucesión de segmentos?

**Cuestión:** ¿Qué significa que hagamos el desarrollo de Taylor de una función $f(x)$? ¿Cómo se relaciona con la cuestión anterior?

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

# ax is a Axes3D object
xs = np.linspace(-3.*np.pi, 3.*np.pi, 60)
ys = np.linspace(-3.*np.pi, 3.*np.pi, 60)
xms, yms = np.meshgrid(xs, ys)
rms = np.sqrt(xms ** 2 + yms ** 2)
zms = np.sin(rms)/rms

# cmap = plt.cm.hot
# other colormaps: plt.cm.jet, plt.cm.warmhot, plt.cm.Blues, plt.cm.plasma
cmap = plt.cm.jet
ct = ax.contour(xms, yms, zms, zdir='z', offset=1.2, alpha=0.4, cmap=cmap)
# ct = ax.contour(mxs, mys, mzs, zdir='x', offset=-5, alpha=0.4, cmap=cmap)
# ct = ax.contour(mxs, mys, mzs, zdir='y', offset=4, alpha=0.4, cmap=cmap)
# rstride are the width or the row, cstride the width of the columns (in units!)
sf = ax.plot_surface(xms, yms, zms, rstride=1, cstride=1, alpha=1., cmap=cmap)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_zlim(-0.3, 1.2);
ax.set_title('$\cos(x^2 + y^2)$')
fig.colorbar(sf, ax=ax);
```

## Parametrización

```python
fig, ax = plt.subplots()
r = 1.
phis = np.linspace(0., 2.*np.pi, 100)
x = r*np.sin(phis)
y = r*np.cos(phis)
ax.plot(x, y, color='blue', lw=2)
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$');
```

**Cuestión:** ¿Cómo puedes dar los puntos del plano que están en la elipse? ¿Cómo harías para recorrer todos los puntos de la circunferencia?

```python
fig, ax = plt.subplots()
x0, y0 = 0., 0.
a, b = 0.5, 3.
r = 1.
phis = np.linspace(0., 2.*np.pi, 100)
xs = x0+a*np.sin(phis)
ys = y0+b*np.cos(phis)
ax.plot(xs, ys, color='blue', lw=2)
ax.set_ylim([np.min(ys)-0.1, np.max(ys)+0.1])
ax.set_xlim([np.min(xs)-0.1, np.max(xs)+0.1])
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$');
```

¿Qué es esta gráfica? ¿Cómo está parametrizada? Modifica los parámetros e interpreta los cambios

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
t = np.linspace(-2., 2., 200)
v, r = 1., 2.
z = v*t
x = r*np.sin(2.*np.pi*t)
y = r*np.cos(2.*np.pi*t)
ax.plot(x, y, z)
ax.set_xlim([-3., 3.])
ax.set_ylim([-3., 3.])
ax.set_zlim([-2., 2.])
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$z$');
ax.set_title(r'$(r sin(2\pi\,t), r \cos(2 \pi \, t), v \, t)$');
```
